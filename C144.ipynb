```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('opencv.png')
n=5
kernel = np.ones((n,n),np.float32)/(n*n)
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
```

```python
img = cv2.imread('check.jpg')

n=11 
#kernel = np.ones((n,n),np.float32)/(n*n)
#dst = cv2.filter2D(img,-1,kernel)

blur = cv2.blur(img,(10,10)) 
#-------------------------------------------
GaussianBlur = cv2.GaussianBlur(img,(n,n),0) #

medianBlur = cv2.medianBlur(img,n)




plt.rcParams["figure.figsize"]=(20,10)
plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(GaussianBlur),plt.title('GaussianBlur')
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(medianBlur),plt.title('medianBlur')
plt.xticks([]), plt.yticks([])
plt.show()
```
