```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
n=10
img = cv2.imread('image5.jpg',0)
img = cv2.bitwise_not(img)
kernel = np.ones((n,n),np.uint8) #사용한 커널의크기에 따라 오브젝트 외곽에사
#0이되는 픽셀의 정도가 달라짐.
erosion = cv2.erode(img,kernel,iterations =2)# 노이즈를 제거하거나 붙어있는 오브젝트들을 분리하는데 사용
#바이너리 이미지에서 흰ㅅ색 오브젝트의 외곽피겔을 0(검은색)으로 만든다.
dilation = cv2.dilate(img,kernel,iterations = 1) # erosion와 반대로 동작
#바이너리 이미지에서 흰색 오브젝트의 외곽 필셀 주변에 1추가
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel) #erosion연산 다음 dilation연산 적용
#이미지상의 노이즈(작은 흰색 물체)를 제거한는데 사용
#노이즈를 없애기위해 사용한 erosion에 의해 작아졌던 오브젯트에 dilation를 적용하면 오브젝트가 원래크기로 돌아옴
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)#opening과 반대로 dilation연산을 먼저 적용한 후 
#erosion연산 적용, 회색 오브젝트에 잇ㅆ는 작은 검은색 구멍들은 메우는데 사용 

gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel) #엣지 검출
```

```python
plt.rcParams["figure.figsize"]=(20,10)
plt.subplot(231),plt.imshow(img),plt.title('Original')
plt.subplot(232),plt.imshow(erosion),plt.title('erosion')
plt.subplot(233),plt.imshow(dilation),plt.title('dilation')
plt.subplot(234),plt.imshow(opening),plt.title('opening')
plt.subplot(235),plt.imshow(closing),plt.title('closing')
plt.subplot(236),plt.imshow(gradient),plt.title('gradient')
#plt.subplot(236),plt.imshow(kernel),plt.title('kernel')
plt.xticks([]), plt.yticks([])
```
