## Exercises 1.4.7 Canny Edge Detection

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.png',0)
blur = cv2.GaussianBlur(img,(5,5),0)
edges = cv2.Canny(img,50,255) #(min,max)
#(200,300)이 값은 0-255사이여야함/ 이 값의 앞 숫자가 커질수록 큰것만 살아남는다.
#값이 작으면 잔잔한것들도 다 살아있음(선같은거)/ (잎숫자가 작고 뒤에 숫자가 커야함)
#max랑 붙어있는 놈은 살리고 떨어져있으면 죽인다. 

#외각부분 찾기
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
 #외곽선 부분 찾아짐-캐니 엣지
    #blur = cv2.GaussianBlur(img,(5,5),0)
    
    #캐니엣지 중요
```
